```scala

val r = requests.get("https://medium.com/netflix-techblog/open-sourcing-polynote-an-ide-inspired-polyglot-notebook-7f929d3f447")
```

```scala
import spark.implicits._
import java.lang.Character
val textDf = Seq(r.data.text).toDF.as[String]
val counts = textDf.flatMap(_.split(" "))
    .filter(s => s.toCharArray.forall(Character.isLetter) && s.size >= 4)
    .map(_.toLowerCase)
    .groupBy('value)
    .count
    .orderBy('count.desc)
    .map(r => r.getString(0) + "," + r.getLong(1))
    .collectAsList
```

```scala
import pandas as pd

df = pd.DataFrame(counts)
df['word'], df['count'] = df[0].str.split(',', 1).str
df['count'] = df['count'].astype(int)
```

```scala
import matplotlib.pyplot as plt

plt.bar(df['word'][0:10], df['count'][0:10], width=0.5, align='center')
plt.xticks(df['word'][0:10], rotation=30)

plt.show()
```
